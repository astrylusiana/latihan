## 1. Turunan

```python
import sympy as sy
```

```python
x = sy.Symbol("x")
y = x**2 + 2*x + 1
```

```python
y
```

```python
sy.diff(y).diff().diff() #turunan ketiga
```

```python
import numpy as np
```

```python
x = np.linspace(-20,20)# dites nilai x nya dari -20 sampe 20, karena numerik tidak bisa membaca fungsi secara langsung
y = x**2 + 2*x + 1
```

```python
x
```

```python
y
```

```python
np.diff(y)/np.diff(x)
```
